```python
import nltk
import json
from database.schema import annotation

def parseAnn(annTxt, lawNum):
    myAnn = annotation()
    for entity in annTxt.split("\n"):
        if len(entity) > 0 and entity[0] == "T":
            obj = entity.split("\t")

            entitySpecs = obj[1].split(" ")
            
            if obj[1].find(";") > 0:
                #print "Found ; seperated location in:", entity
                continue
            
            entity = entitySpecs[0]
            start = int(entitySpecs[1])
            end = int(entitySpecs[2])

            phrase = obj[2]

            myAnn.entityType = entity
            myAnn.start = start
            myAnn.end = end
            myAnn.phrase = phrase
            myAnn.lawNum = lawNum
            ann_db.insert_one(myAnn.out())
            
            print start, end, entity, phrase
        


from pymongo import MongoClient
import glob
import os

path = '/home/mlc/BobGoesToJail/corpus/lawTexts_en'

client = MongoClient()
client = MongoClient('localhost', 27017)
db = client['law_db']
ann_db = db.anns

for filename in glob.glob(os.path.join(path, '*.ann')):
    print filename
    fAnn = open(filename)
    lawNum = filename[filename.rfind('/') + 1:filename.rfind('.')]
    parseAnn(fAnn.read(), lawNum)   
    fAnn.close()     
```
